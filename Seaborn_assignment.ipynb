## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Ans:

Seaborn is a popular data visualization library in Python that builds on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common types of plots we can create using Seaborn, along with their typical use cases:

1. Histograms (sns.histplot()):\
        Use: Histograms are used to visualize the distribution of a single numerical variable. They display the frequency or count of data points falling within specified bins or intervals.\
        Example: We can use histograms to visualize the distribution of exam scores in a class to understand how the scores are distributed across different grade ranges.

2. Scatter Plots (sns.scatterplot()):\
        Use: Scatter plots are used to visualize the relationship between two numerical variables. Each data point is represented as a dot on the plot, allowing you to see patterns, trends, or correlations.\
        Example: We can create scatter plots to explore the relationship between a person's age and their income to determine if there's a correlation between these two variables.

3. Bar Plots (sns.barplot()):\
        Use: Bar plots are used to display the relationship between a categorical variable and a numerical variable. They show the average value of the numerical variable for each category.\
        Example: We can use bar plots to compare the average sales revenue for different product categories in a store.

4. Box Plots (sns.boxplot()):\
        Use: Box plots are used to visualize the distribution of a numerical variable across different categories or groups. They display the quartiles, median, and potential outliers of the data.\
        Example: We can create box plots to compare the distribution of salaries for employees in different departments of a company to identify variations.

5. Heatmaps (sns.heatmap()):\
        Use: Heatmaps are used to visualize the pairwise relationships and correlations between variables in a dataset. They use color intensity to represent the strength and direction of correlations.\
        Example: We can use a heatmap to visualize the correlation matrix of stock prices for a portfolio of assets to identify which assets are positively or negatively correlated.

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

## Ans:

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Create box plots for 'age' and 'fare' with respect to 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Box Plot of Age by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

## Ans:

```python
diamond = sns.load_dataset('diamonds')
diamond
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(diamond['price'], bins=30, kde=True, color='blue')
# Set labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset('diamonds')

# Create a histogram for the 'price' column with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', bins=30, kde=True, hue='cut', palette='Set1')

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices by Cut')

# Show the legend
plt.legend(set(diamonds_data['cut']))

# Show the plot
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

## Ans:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset('iris')

# Create a pair plot with hue for the "species" column
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])

# Show the plot
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

## Ans:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset('flights')

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Counts by Month and Year')

# Show the plot
plt.show()
```
