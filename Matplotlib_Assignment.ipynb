## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:\
import numpy as np\
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500]) \
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])\
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])\
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

## Ans:

The subplot() function in most programming libraries, such as Matplotlib in Python, is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots or charts in a structured grid, making it easier to compare and visualize different data or aspects of your data in the same figure. Each subplot is like a smaller canvas within the main figure, and you can specify its position and dimensions within the grid using row and column indices.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x, y1)
plt.title('1st plot')

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x, y2)
plt.title('2nd plot')

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x, y3)
plt.title('3rd plot')

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x, y4)
plt.title('4th plot')

# Adjust layout
plt.tight_layout()# Without this titles,x-label and y-labels starts overlaping

# Show the figure
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

## Ans:

A bar plot, also known as a bar chart or bar graph, is a graphical representation of data that uses rectangular bars or columns to represent different categories or data points. Bar plots are widely used in data visualization and have several purposes:

1. Comparison of Data: Bar plots are used to compare different categories or data points. Each bar represents a category, and the height or length of the bar is proportional to the value or frequency of that category. This makes it easy to visually compare the magnitude of different data points.

2. Categorical Data: Bar plots are especially useful for displaying categorical data or data that can be divided into distinct categories. These categories can be anything from product names, months, countries, to survey responses or any discrete data.

3. Visualizing Distribution: Bar plots can also be used to visualize the distribution of data within categories. For example, you can create a histogram-like bar plot to see how data is distributed across different bins or intervals.

4. Showing Trends: Bar plots can show trends over time or across categories. For example, you can create a bar plot to display monthly sales figures for a product over a year, making it easy to identify seasonal patterns or trends.

5. Comparison of Multiple Variables: Stacked bar plots or grouped bar plots can be used to compare multiple variables within each category. This is useful for visualizing how different factors contribute to the overall value in each category.

```python
plt.bar(company,profit)
plt.xlabel('Names of the company')
plt.ylabel('Profit earned')
plt.title('Bar plot of company profit')
plt.grid()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

## Ans:

A box plot, also known as a whisker plot or box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is used to display the summary statistics and key characteristics of a dataset, providing a visual summary of its central tendency, spread, and potential outliers. Box plots are valuable in data analysis and visualization for several reasons:

1. Summary of Data Distribution: A box plot provides a concise summary of the distribution of a dataset, including the median, quartiles (25th and 75th percentiles), and potential outliers. This summary allows you to quickly grasp the central tendency and spread of the data.

2. Identification of Outliers: Box plots can easily identify outliers in the data. Outliers are data points that fall significantly above or below the "whiskers" (the lines extending from the box). They are represented as individual points beyond the whiskers, making them visually distinct.

3. Comparison of Distributions: Box plots are useful for comparing the distributions of different datasets or categories. When multiple box plots are displayed side by side, it becomes easy to compare their central tendencies, spreads, and outlier characteristics.

4. Visualizing Skewness: The asymmetry of a dataset (skewness) can be visualized using box plots. If one whisker is longer than the other, it indicates that the data is skewed in that direction.

5. Detection of Data Spread: The length of the box in a box plot represents the interquartile range (IQR), which gives a sense of the data's spread. A wider box indicates greater variability, while a narrower box suggests less variability.

6. Comparing Groups: Box plots are effective for comparing multiple groups or categories within a single variable. They help in understanding the variation between groups and identifying potential differences or patterns.

7. Robustness to Outliers: Unlike some other plots, box plots are robust to extreme outliers. They don't extend all the way to the minimum and maximum values but only up to a certain range, making them less affected by extreme values.

```python
# Determine the common y-axis limits
y_min = min(np.min(box1), np.min(box2))
y_max = max(np.max(box1), np.max(box2))

# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(8, 4))

# Create box plots for each dataset in the same subplot
ax.boxplot([box1, box2],vert=False)

# Set the same y-axis limits for both box plots
ax.set_xlim(y_min - 1, y_max + 1)

# Set labels for the box plots
ax.set_yticklabels(['Data 1', 'Data 2'])

# Set a title
ax.set_title('Box Plots with Same Scale')

# Show the figure
plt.show()
```
