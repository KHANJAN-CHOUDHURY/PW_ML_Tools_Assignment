## Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

## Ans:

```python
!pip install plotly
```

```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
fig = go.Figure()
fig.update_layout(title="Age VS Fare",xaxis_title = 'Age',yaxis_title = 'Fare')
fig.add_trace(go.Scatter(x = titanic.age, y = titanic.fare, mode = 'markers'))
```

## Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

## Ans:

```python
import plotly.express as px
df = px.data.tips()
df
```

```python
fig = px.box(df, y="total_bill")
fig.show()
```

## Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

## Ans:

```python
fig1 = px.histogram(
    df,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Total Bill by Sex with Day and Smoker Information",)
fig1.show()
```

## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

## Ans:

```python
iris = px.data.iris()

# Create the scatter matrix plot
fig2 = px.scatter_matrix(
    iris,
    dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset by Species",
)

# Show the plot
fig2.show()
```

## Q5. What is Distplot? Using Plotly express, plot a distplot.

## Ans:

A "distplot" is a distribution plot used to visualize the distribution of a univariate dataset. It combines elements of a histogram and a kernel density estimate (KDE) plot to provide a comprehensive view of the data's distribution, including its central tendency and spread. Distplots are commonly used to explore the shape of a dataset and identify patterns such as bimodality or skewness.

To create a distplot using Plotly Express, you can use the px.histogram function with the marginal parameter set to "rug" or "box" to display a KDE curve or a box plot along with the histogram. Here's an example:

```python
import plotly.express as px

# Sample data (replace with your own dataset)
data = titanic.fare

# Create a distplot
fig = px.histogram(data, x=data, marginal="box", title="Distplot Example")

# Show the plot
fig.show()
```
